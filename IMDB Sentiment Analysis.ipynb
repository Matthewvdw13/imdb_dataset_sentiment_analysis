```python
from keras.datasets import imdb
from keras.models import Sequential
from keras.layers import Dense, GRU, Flatten, LSTM  
from keras.layers.embeddings import Embedding
from keras.preprocessing import sequence

import pandas as pd
import numpy as np
```

```python
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words = 5000)
```

```python
X_train.shape
```

```python
X_train = sequence.pad_sequences(X_train, maxlen=500)
X_test = sequence.pad_sequences(X_test, maxlen=500)
```

```python
model = Sequential()
model.add(Embedding(5000, 100, input_length=500))
model.add(GRU(100))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, epochs=3, batch_size=64)
```

```python
scores = model.evaluate(X_test, y_test)
```

```python
print("The model scored an accuracy of on the test set: %.2f%%" % (scores[1]*100))
```

```python
y_pred = model.predict(X_test) #This variable stores all the predicted integer values for the reviews
```

```python
word_index = imdb.get_word_index()
```

I decided to illustrate the real values and predicted values as integers instead of Positive and Negative because the integer values give you far more insight to how close the model was to the real value. 

```python
def decode_review(x):
    print(x)
    review_text = ' '.join([dict([(value, key) for (key, value) in word_index.items()]).get(i - 3, '?') for i in x])
    print(review_text)
```

```python
for x, y, z in zip(X_test[:10], y_test[:10], y_pred[:10]):
    decode_review(x)
    print("\n")
    print("Real Value:")
    print(y)
    print("\n")
    print("Predicted Value:")
    print(z)
    print("\n")
```
